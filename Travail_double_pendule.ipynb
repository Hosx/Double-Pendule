# Présentation du problème

Nous allons étudier ici le problème du double pendule comme représenté sur le schéma ci-dessous.\\\\

Les positions des centres de masses des deux barres sont données par
$$\begin{cases}
    x_1 = \displaystyle{\frac{l}{2} \sin{\theta_1}}\\
    y_1 = -\displaystyle{\frac{l}{2} \cos{\theta_1}}
\end{cases}$$
et 
$$\begin{cases}
    x_2 = \displaystyle{l( \sin{\theta_1} + \frac{1}{2} \sin{\theta_2}})\\
    y_2 = \displaystyle{-l(\cos{\theta_1} + \frac{1}{2} \cos{\theta_2})}
\end{cases}$$
On définit le **Lagrangien** du système comme étant $\mathcal{L} = T - V$ où $T$ est l'énergie cinétique et $V$ le potentiel. Explicitons ce Lagrangien. On sait que pour une rotation, nous avons
\begin{equation}
    T = \frac{1}{2} m v^2 + \frac{1}{2} I \omega^2
\end{equation}

où $I$ est l'inertie de la barre par rapport à l'origine $O$ et $\omega$ est la vitesse angulaire. On sait que l'inertie est donnée par
\begin{equation}
I = \int_{-l/2}^{l/2} x^2 \rho ~dx = \frac{1}{12}\rho l^3 = \frac{1}{12} m l^2
\end{equation}
et $\omega^2 = \dot{\theta}_1^2 +  \dot{\theta}_2^2$. Nous avons donc déjà explicité le second terme de l'énergie cinétique, voyons le premier. On sait que 
\begin{align*}
    v^2 &= \dot{x}_1^2 + \dot{y}_1^2 + \dot{x}_2^2 + \dot{y}_2^2\\
    &= \frac{l^2}{4}(\dot{\theta}_1^2 \cos^2{\theta_1} + \dot{\theta}_1^2 \sin^2{\theta_1}) + l^2\left(\dot{\theta}_1\cos{\theta_1} + \frac{\dot{\theta}_2}{2}\cos{\theta_2}\right)^2 + l^2\left(\dot{\theta}_1\sin{\theta_1} + \frac{\dot{\theta}_2}{2}\sin{\theta_2}\right)^2 \\
    &= \frac{l^2}{4}\dot{\theta}_1^2 + l^2\left(\dot{\theta}^2_1\cos^2{\theta_1} + \frac{\dot{\theta}_2^2}{4} \cos^2{\theta_2} + \dot{\theta}_1\dot{\theta}_2\cos{\theta_1}\cos{\theta_2}\right) \\  
    &~~~  + l^2\left(\dot{\theta}^2_1\sin^2{\theta_1} + \frac{\dot{\theta}_2^2}{4} \sin^2{\theta_2} + \dot{\theta}_1\dot{\theta}_2\sin{\theta_1}\sin{\theta_2}\right)\\
    &= \frac{l^2 \dot{\theta}_1^2}{4} + l^2 \dot{\theta}_1^2 + \frac{l^2 \dot{\theta}_2^2}{4} + l^2 \dot{\theta}_1\dot{\theta}_2 \cos(\theta_1 - \theta_2)
\end{align*}
On trouve alors 
\begin{align*}
    T &= \frac{1}{2} m l^2 \left(\frac{5 \dot{\theta}_1^2}{4} + \frac{\dot{\theta}_2^2}{4} + \dot{\theta}_1\dot{\theta}_2 \cos(\theta_1 - \theta_2)\right) + \frac{1}{24} ml^2(\dot{\theta}_1^2 + \dot{\theta}_2^2)\\
    &= \frac{ml^2}{6}(4\dot{\theta}_1^2 + \dot{\theta}_2^2 + 3\dot{\theta_1}\dot{\theta_2}\cos{\theta_1 - \theta_2})
\end{align*}

De plus, l'énergie potentielle est donnée par
$$V = mg(y_1+y_2) = -mgl\left(\frac{1}{2} \cos{\theta_1} + \cos{\theta_1} + \frac{1}{2} \cos{\theta_2}\right) = -\frac{mgl}{2}\left( 3\cos{\theta_1} + \cos{\theta_2}\right)$$
On trouve dans ce cas, 
$$\mathcal{L} =\frac{ml^2}{6}(4\dot{\theta}_1^2 + \dot{\theta}_2^2 + 3\dot{\theta_1}\dot{\theta_2}\cos{\theta_1 - \theta_2})+ \frac{mgl}{2}\left( 3\cos{\theta_1} + \cos{\theta_2}\right)$$

Construisons maintenant les **équations de Hamilton-Jacobi**. Résoudre ce système d'équation est équivalent à résoudre les équation Lagrangienne, elles nous permettrons donc de décrire le mouvement du double pendule. Nous définissons pour ce faire les *impulsions généralisées* telles que
\begin{equation*}
\begin{cases}
    p_{\theta_1} = \displaystyle{\frac{\partial\mathcal{L}}{\partial \dot{\theta}_1}}\\
    \\
    p_{\theta_2} = \displaystyle{\frac{\partial\mathcal{L}}{\partial\dot{\theta}_2}}
\end{cases}
\end{equation*}
Ces équations donnent dans ce cas-ci
\begin{equation*}
\begin{cases}
    p_{\theta_1} = \displaystyle{\frac{1}{6}ml^2(8\dot{\theta}_1 + 3\dot{\theta}_2\cos(\theta_1 - \theta_2))}\\
    \\
    p_{\theta_2} = \displaystyle{\frac{1}{6}ml^2(2\dot{\theta}_2 + 3\dot{\theta}_1\cos(\theta_1 - \theta_2))}
\end{cases}
\end{equation*}

Remarquons que ces équations sont inversibles et que nous pouvons isoler $\dot{\theta}_1$ et $\dot{\theta}_2$. En combinant les deux équations ci-dessus, nous obtenons alors
\begin{equation*}
\begin{cases}
    \dot{\theta}_1 = \displaystyle{\frac{6}{ml^2}\frac{2p_{\theta_1} - 3\cos(\theta_1 - \theta_2)p_{\theta_2}}{16 - 9\cos^2(\theta_1 - \theta_2)}}\\
    \\
    \dot{\theta}_2 = \displaystyle{\frac{6}{ml^2}\frac{8p_{\theta_2} - 3\cos(\theta_1 - \theta_2)p_{\theta_1}}{16 - 9\cos^2(\theta_1 - \theta_2)}}
\end{cases}
\end{equation*}

Les équations de Hamilton-Jacobi se résument à 
\begin{equation*}
\begin{cases}
    \dot{q}^{\alpha} = \displaystyle{\frac{\partial\mathcal{H}}{\partial p_{\alpha}}}\\
    \\
    \dot{p}_{\alpha} = -\displaystyle{\frac{\partial\mathcal{H}}{\partial q^{\alpha}}}
\end{cases}
\end{equation*}

où $q^{\alpha}$ désigne $\theta_1$ ou $\theta_2$, $p_{\alpha}$ désigne $p_{\theta_1}$ ou $p_{\theta_2}$ et $\mathcal{H}(q^{\alpha}, p_{\alpha}) = p_{\theta_1} \dot{\theta}_1(p_{\theta_1}, p_{\theta_2}) + p_{\theta_2} \dot{\theta}_2(p_{\theta_1}, p_{\theta_2})  -  \mathcal{L}(q^{\alpha}, p_{\alpha})$. Nous pourrions calculer l'hamiltonien explicitement pour déduire les équations de Hamilton-Jacobi explicitement, seulement nous pouvons simplifier le raisonnement. En effet, pour la première équation, nous l'avons explicité précédemment dans notre cas particulier (mais ceci est tout à fait général à tout lagrangien et hamiltonien), nous avons que les relations 
$$\dot{q}^{\alpha} = \displaystyle{\frac{\partial\mathcal{H}}{\partial p_{\alpha}}} \qquad \text{et} \qquad p_{\alpha} = \displaystyle{\frac{\partial\mathcal{L}}{\partial \dot{q}^{\alpha}}}$$
sont inverses l'une de l'autre, donc si nous avons la première nous pouvons trouver la seconde et inversément en isolant les variables. Pour la seconde équation, nous avons vu au cours de mécanique analytique (nous vous épargnons le développement de ceci) que 
$$\frac{\partial\mathcal{H}}{\partial q^{\alpha}} = -\frac{\partial\mathcal{L}}{\partial q^{\alpha}}$$

Avec ceci, nous en déduisons que la connaissance du lagrangien suffit pour écrire les équations d'Hamilton-Jacobi, il n'est pas nécessaire d'explicité l'hamiltonien du système. Les équations sont alors
\begin{equation*}
\begin{cases}
    \dot{\theta}_1 = \displaystyle{\frac{6}{ml^2}\frac{2p_{\theta_1} - 3\cos(\theta_1 - \theta_2)p_{\theta_2}}{16 - 9\cos^2(\theta_1 - \theta_2)}}\\
    \\
    \dot{\theta}_2 = \displaystyle{\frac{6}{ml^2}\frac{8p_{\theta_2} - 3\cos(\theta_1 - \theta_2)p_{\theta_1}}{16 - 9\cos^2(\theta_1 - \theta_2)}}\\
    \\
    \dot{p}_{\theta_1} = \displaystyle{\frac{\partial\mathcal{L}}{\partial \theta_1} = \frac{-1}{2} ml^2 \left(\dot{\theta}_1\dot{\theta}_1 \sin(\theta_1 - \theta_2) + \frac{3g}{l} \sin{\theta_1}\right)}\\
    \\
    \dot{p}_{\theta_2} = \displaystyle{\frac{\partial\mathcal{L}}{\partial \theta_2} = \frac{-1}{2} ml^2 \left(-\dot{\theta}_1\dot{\theta}_1 \sin(\theta_1 - \theta_2) + \frac{g}{l} \sin{\theta_2}\right)}
\end{cases}
\end{equation*}

Posons alors 
\begin{equation*}
    y = 
    \begin{pmatrix}
    \theta_1\\
    \theta_2\\
    p_{\theta_1}\\
    p_{\theta_2}
    \end{pmatrix}
\end{equation*}

Ce qui nous donne
\begin{equation*}
    \dot{y} = 
    \begin{pmatrix}
    \dot{\theta}_1\\
    \dot{\theta}_2\\
    \dot{p}_{\theta_1}\\
    \dot{p}_{\theta_2}
    \end{pmatrix} =
    f(y, \dot{y}) =
    \begin{pmatrix}
    \displaystyle{\frac{6}{ml^2}\frac{2p_{\theta_1} - 3\cos(\theta_1 - \theta_2)p_{\theta_2}}{16 - 9\cos^2(\theta_1 - \theta_2)}}\\
    \\
    \displaystyle{\frac{6}{ml^2}\frac{8p_{\theta_2} - 3\cos(\theta_1 - \theta_2)p_{\theta_1}}{16 - 9\cos^2(\theta_1 - \theta_2)}}\\
    \\
    \displaystyle{ \frac{-1}{2} ml^2 \left(\dot{\theta}_1\dot{\theta}_1 \sin(\theta_1 - \theta_2) + \frac{3g}{l} \sin{\theta_1}\right)}\\
    \\
    \displaystyle{\frac{-1}{2} ml^2 \left(-\dot{\theta}_1\dot{\theta}_1 \sin(\theta_1 - \theta_2) + \frac{g}{l} \sin{\theta_2}\right)}
    \end{pmatrix} 
\end{equation*}
Qui est une équation différentielle du premier ordre que nous somme en mesure de résoudre à l'aide de l'algorithme rk4.


# Résolution algorithmique
Nous allons maintenant résoudre les équations différentielles ci-dessus à l'aide de l'algorithme de Runge-Kutta d'ordre 4.

```python
from numpy import sin, cos
from scipy import *
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import matplotlib.animation as animation
from IPython.display import HTML

#On introduit les constantes nécessaire à la résolution du problème.
g = 9.81 #gravité
l = 1 #longueur des barreaux
m = 1 #masse des barreaux
T = 20 #temps final auquel on voudrait arrivé
dt = 0.001 #pas entre chaque étape de calcul
yn1 = np.array([math.pi/2, math.pi/2, 0, 0]) #condition initiale de notre problème

# On définit la fonction f explicitée précédemment qui prend comme valeur le vecteur [theta1, theta2, p_theta1, p_theta2] et
# qui renvoit le vecteur [dtheta1, dtheta2, dp_theta1, dp_theta2], où le "d" désigne la dérivée, ce vecteur est explicité ci-dessus.
def f(y):
    cos = math.cos(y[0] - y[1]) #Cette expression revient souvent, mieux vaut la calculer qu'une seule fois
    sin = math.sin(y[0] - y[1]) 
    dtheta_1 = 6*(2*y[2] - 3*cos*y[3])/((16- 9*cos**2)*m*l**2) #Exprime la dérivée de theta1 
    dtheta_2 = 6*(8*y[3] - 3*cos*y[2])/((16- 9*cos**2)*m*l**2) #Exprime la dérivée de theta2 
    dp_th1 = -0.5*m*l**2*(dtheta_1*dtheta_2*sin + 3*g*math.sin(y[0])/l) #Exprime la dérivée de p_theta1 
    dp_th2 = -0.5*m*l**2*(-dtheta_1*dtheta_2*sin + g*math.sin(y[1])/l) #Exprime la dérivée de p_theta2 
    f = np.array([dtheta_1, dtheta_2, dp_th1, dp_th2]) #Est le vecteur renvoyé par f explicité ci-dessus
    return f

    #f1 = 6/(m*l**2)*(2*p1-3*np.cos(theta1-theta2)*p2)/(16-9*np.cos(theta1-theta2)**2)
    #f2 = 6/(m*l**2)*(8*p2-3*np.cos(theta1-theta2)*p1)/(16-9*np.cos(theta1-theta2)**2)
    f3 = -0.5*m*l**2*(f1*f2*np.sin(theta1-theta2)+3*g/l*np.sin(theta1))
    f4 = -0.5*m*l**2*(-f1*f2*np.sin(theta1-theta2)+g/l*np.sin(theta2))


# Algorithme de Runge Kutta d'ordre 4, elle prend comme argument: la condition initiale yn, le temps T durant lequel on voudrait
# faire la mesure et l'écart de temps dt entre chaque étape. Elle nous renvoit une liste avec les valeurs de 
# [theta1, theta2, p_theta1, p_theta2] à chaque étape dt.

def rk4(dt, T, yn):
    N = int(T / dt) + 1 #Nous donne le nombre d'étapes à effectuer entre 0 et T par pas dt.
    a=[yn]
    for n in range(1, N): #boucle qui, à chaque étape, va effectuer la méthode Runge Kutta
        k1 = dt*f(yn) #définit k1 de Runge Kutta
        k2 = dt*f(yn + k1/2) #définit k2 de Runge Kutta
        k3 = dt*f(yn + k2/2) #définit k3 de Runge Kutta
        k4 = dt*f(yn + k3) #définit k4 de Runge Kutta
        yn = yn + (k1 + 2*k2 + 2*k3 + k4)/6 #Nous donne [theta1, theta2, p_theta1, p_theta2] après un pas dt
        a.append(yn)
    return np.array(a)


RK4_yn=rk4(dt,T,yn1)


```

    \end{bmatrix}
$$

# Convergence
Cette partie du travail est très similaire à celle réalisée dans le MOOC concernant la grille de convergence et l'ordre de convergence dans "Full Phugoïd Model". \\
N'ayant pas de solution exact pour notre équation différentielle, nous allons comparer les solutions obtenues pour différentes valeurs de notre pas $dt$. Nous considérerons alors notre plus petite valeur de $dt$ comme étant la solution exacte et nous appliquerons ensuite la norme $L^1$ sur les autres valeurs plus larges de notre pas. C'est à dire

$$\int_0^T |y_n(t) - y(t)|~dt$$

où $y$ est notre solution pour le plus petit pas $dt$ et $y_n$ celle pour les autres pas. Naturellement, en informatique nous ne travaillons pas avec des intégrales. Néanmoins, nous pouvons approximer l'équation ci dessus par

$$dt\sum_{n = 0}^N |y_n(ndt) - y(ndt)|$$

où $dt$ est le pas de la solution la plus grossière et $N$ est tel que $Ndt = T$. L'ensemble des valeurs de ceci pour les différentes valeurs de $dt$ est ensuite représenté graphiquement en fonction de $dt$.

```python
# Ensemble des n différentes valeurs de dt que nous allons considérer.
valeurs_dt = [0.1, 0.01, 0.005, 0.001, 0.0005, 0.00025]

# On crée une liste vide qui va contenir les n listes (une pour chacune des valeurs de dt) avec toutes les solutions y données 
# par rk4 à ce dt.
valeurs_y = []

# On ajoute les solutions y pour chacun des dt à notre liste valeurs_y
for dt in valeurs_dt:
    valeurs_y.append(rk4(dt, T, yn1))

    
# On crée une fonction L1_diff qui prend comme argument une solution y grossière (grand dt) et une solution y plus fine (le plus
# petit dt) ainsi que le dt de la solution grossière. Elle nous renvoit ensuite la valeur de la norme L1.
# Il y a néanmoins une chose à prendre en compte ici, c'est que le nombre d'élément dans y_precis est plus grand que dans y_grossier
# donc si nous voulons pouvoir faire des opérations sur ces listes, il faudrait qu'elles soient de même taille. Pour ce faire, 
# nous découpons la liste y_precis (comme nous l'avons vu dans le MOOC) de manière à ce que les deux listes soient de même tailles.
def l1_diff(y_grossier, y_precis, dt):
    N_grossier = len(y_grossier) #Longueur de y_grossier
    N_precis = len(y_precis) #Longueur de y_precis
    ratio = math.ceil(N_precis / N_grossier) #Rapport des longueurs des deux listes
    diff = dt * np.sum(np.abs(y_grossier - y_precis[::ratio])) # Norme L1 pour des listes de même tailles
    
    return diff

# On crée une liste vide qui va contenir les n-1 normes L1 pour une des variable de y = [theta1, theta2, p_theta1, p_theta2]
# Nous avons pris ici comme variable p_theta2
valeurs_diff = []

# Pour chacune des valeurs de dt différente de la plus fine on calcul la norme L1 sur la 3e variable de y (p_theta2).
for i, dt in enumerate(valeurs_dt[:-1]):
    diff = l1_diff(valeurs_y[i][:,2], valeurs_y[-1][:,2], dt) #Calcul la norme L1 entre y plus précis et y grossière
    valeurs_diff.append(diff) #On ajoute notre valeur à notre liste valeurs_diff
```

## Ordre de convergence
Nous prenons ici trois valeurs de $dt$ donnée par 0.00025, 0.0005 et 0.001 qui sont chacune des multiples de 2 de la plus petite valeur de $dt$. Nous avons vu dans le cours que l'ordre de convergence est donné par
$$p = \frac{\log\left(\frac{y_3 - y_2}{y_2 - y_1}\right)}{\log(r)}$$

Dans le cas présent, ces différences de $y$ sont précisément les $L^1$-norme calculée précédemment (pour la variable $p_{\theta_1}$). Etant donné que nous travaillons avec Runge-Kutta d'ordre 4, il semble clair que l'ordre de convergence doit être de l'ordre de 4, ce qui est bien la valeur obtenue.
